```python
import pandas as pd
from sklearn import tree
from sklearn .tree import DecisionTreeClassifier
import numpy as np
from sklearn.utils import shuffle
```

```python
df = pd.read_excel('C题.xlsx')
df = df.dropna(how = 'any',axis = 0)
df = shuffle(df, random_state=42) 
df = df.reset_index(drop = True)
df
```

```python
idx = 0
for i in df.loc[:,'纹饰']:
    if i == 'A':
        df.loc[idx,'纹饰'] = 0
    idx += 1
df
idx = 0
for i in df.loc[:,'纹饰']:
    if i == 'B':
        df.loc[idx,'纹饰'] = 1
    idx += 1
df
idx = 0
for i in df.loc[:,'纹饰']:
    if i == 'C':
        df.loc[idx,'纹饰'] = 2
    idx += 1
df
idx = 0
for i in df.loc[:,'类型']:
    if i == '高钾':
        df.loc[idx,'类型'] = 3
    idx += 1
df
idx = 0
for i in df.loc[:,'类型']:
    if i == '铅钡':
        df.loc[idx,'类型'] = 4
    idx += 1
df
idx = 0
for i in df.loc[:,'颜色']:
    if i == '蓝绿':
        df.loc[idx,'颜色'] = 0
    idx += 1
df
idx = 0
for i in df.loc[:,'颜色']:
    if i == '深绿':
        df.loc[idx,'颜色'] = 1
    idx += 1
df
idx = 0
for i in df.loc[:,'颜色']:
    if i == '浅绿':
        df.loc[idx,'颜色'] = 2
    idx += 1
df
idx = 0
for i in df.loc[:,'颜色']:
    if i == '深蓝':
        df.loc[idx,'颜色'] = 3
    idx += 1
df
idx = 0
for i in df.loc[:,'颜色']:
    if i == '浅蓝':
        df.loc[idx,'颜色'] = 4
    idx += 1
df
idx = 0
for i in df.loc[:,'颜色']:
    if i == '绿':
        df.loc[idx,'颜色'] = 5
    idx += 1
df
idx = 0
for i in df.loc[:,'颜色']:
    if i == '黑':
        df.loc[idx,'颜色'] = 6
    idx += 1
df
idx = 0
for i in df.loc[:,'颜色']:
    if i == '紫':
        df.loc[idx,'颜色'] = 7
    idx += 1
df
idx = 0
for i in df.loc[:,'表面风化']:
    if i == '无风化':
        df.loc[idx,'表面风化'] = 0
    idx += 1
df
idx = 0
for i in df.loc[:,'表面风化']:
    if i == '风化':
        df.loc[idx,'表面风化'] = 1
    idx += 1
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df.loc[:,['纹饰','类型']],df.loc[:,'表面风化'],test_size=0.2)
clf = DecisionTreeClassifier(criterion="gini",max_depth = 10)
# df.loc[:,'纹饰'] = df.loc[:,'纹饰'].astype('int')
clf = clf.fit(X_train.astype('int'),Y_train.astype('int'))
score = clf.score(X_test, Y_test.astype('int')) #返回预测的准确度
score
```

```python
import graphviz
import matplotlib.pyplot as plt

feature_name = ['类型','纹饰']

data = tree.export_graphviz(clf,out_file=None
                                ,feature_names=feature_name 
                                ,class_names=['风化','不风化'] 
                                ,filled=True
                                ,rounded=True)
graph = graphviz.Source(data)
graph
```
